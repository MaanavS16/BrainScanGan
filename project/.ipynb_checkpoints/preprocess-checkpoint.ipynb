```python
import os
#from pathlib import Path

import matplotlib
from matplotlib import pylab as plt

import nibabel as nib
from nibabel import nifti1
from nibabel.viewers import OrthoSlicer3D

import csv

metadata_path = os.path.join(os.path.dirname(os.getcwd()), "research_dataset", "patient_metadata.csv")
metadata_path
```

```python
patient_metadata = {}
with open(metadata_path) as csvfile:
    reader = csv.reader(csvfile)
    keys = next(reader)
    for row in reader:
        data_list = []
        for i, v in enumerate(keys):
            data_list.append({v:row[i]})
        patient_metadata.update({str(row[0]): data_list})

patient_data_paths = list(filter(lambda x: ".csv" not in x, os.listdir("../research_dataset")))
patient_data_paths = [os.path.join(os.path.dirname(os.getcwd()), "research_dataset", x) for x in patient_data_paths]
#patient_data_paths
```

```python


import numpy as np
from matplotlib import animation

def pathToArray(path):
    image_array = nib.load(path)
    image_array.get_fdata()
    image_array = image_array.get_fdata()
    #image_array.shape
    return image_array


sample_element = pathToArray(os.path.join(patient_data_paths[0], "ses-11300_NoMyeline", "sub-CC00058XX09_ses-11300_desc-restore_T2w.rmcere.nii.gz"))

fig = plt.figure()

i = 0
im = plt.imshow(sample_element[:,:,i])

def updatefig(*args):
    global i
    im.set_array(sample_element[:,:,i])
    i += 1
    return (im,)

ani = animation.FuncAnimation(fig, updatefig, frames=100, interval=100)
#plt.show()
plt.show()

```
